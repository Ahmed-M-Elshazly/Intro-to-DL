```python
import pulp
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import T5Tokenizer, T5ForConditionalGeneration
from transformers import get_linear_schedule_with_warmup
import os

torch.cuda.empty_cache()

# 1) Convex oracle: PuLP optimization for TCL problem
def generate_optimal_schedule(prices, ambient_temperatures, P_AC, T_room_init, T_room_min, T_room_max, alpha, G, peak_power_limit):
    N = len(prices) - 1
    prob = pulp.LpProblem("TCL_Cost_Minimization", pulp.LpMinimize)

    x1 = [pulp.LpVariable(f"x1_{n}", 0, 1) for n in range(N+1)]
    x2 = [pulp.LpVariable(f"x2_{n}", 0, 1) for n in range(N+1)]
    T_room1 = [pulp.LpVariable(f"T_room1_{n}", T_room_min[0], T_room_max[0]) for n in range(N+1)]
    T_room2 = [pulp.LpVariable(f"T_room2_{n}", T_room_min[1], T_room_max[1]) for n in range(N+1)]

    prob += pulp.lpSum([prices[n]*(P_AC[0]*x1[n] + P_AC[1]*x2[n]) for n in range(N+1)])

    prob += T_room1[0] == T_room_init[0]
    prob += T_room2[0] == T_room_init[1]

    for n in range(N):
        prob += T_room1[n+1] == T_room1[n] + alpha[0]*(ambient_temperatures[n] - T_room1[n]) + G[0]*P_AC[0]*x1[n]
        prob += T_room2[n+1] == T_room2[n] + alpha[1]*(ambient_temperatures[n] - T_room2[n]) + G[1]*P_AC[1]*x2[n]

    for n in range(N+1):
        prob += P_AC[0]*x1[n] + P_AC[1]*x2[n] <= peak_power_limit

    status = prob.solve(pulp.PULP_CBC_CMD(msg=False))

    return {
        'x1': np.array([pulp.value(x) for x in x1]),
        'x2': np.array([pulp.value(x) for x in x2]),
        'status': pulp.LpStatus[status]
    }

# 2) Structured Prompt Dataset for T5
class TCLSupervisedDatasetStructured(Dataset):
    def __init__(self, n_samples, N):
        self.prompts = []
        self.targets = []
        self.N = N
        self.P_AC = [2.0, 2.5]  # kW
        self.T_room_init = [25.0, 24.0]
        # Define ranges for min and max temperatures
        self.T_room_min_range = [(20.0, 23.0), (19.0, 22.0)]  # (AC1 range, AC2 range)
        self.T_room_max_range = [(26.0, 29.0), (25.0, 28.0)]  # (AC1 range, AC2 range)
        self.alpha = [0.1, 0.1]
        self.G = [-0.3, -0.3]
        self.peak_range = (3.0, 6.0)

        for _ in range(n_samples):
            prices = 0.05 + (0.2 - 0.05) * np.random.rand(N+1)
            ambient_temperatures = 28 + (35 - 28) * np.random.rand(N+1)
            peak_power_limit = np.random.uniform(*self.peak_range)

            # Randomly sample min and max temperatures for this sample
            T_room_min = [
                np.random.uniform(*self.T_room_min_range[0]),
                np.random.uniform(*self.T_room_min_range[1])
            ]
            T_room_max = [
                np.random.uniform(*self.T_room_max_range[0]),
                np.random.uniform(*self.T_room_max_range[1])
            ]

            optimal = generate_optimal_schedule(
                prices, ambient_temperatures,
                self.P_AC, self.T_room_init,
                T_room_min, T_room_max,
                self.alpha, self.G, peak_power_limit
            )

            prompt = self._build_prompt(prices, ambient_temperatures, peak_power_limit, T_room_min, T_room_max)
            target = self._schedules_to_text(optimal['x1'], optimal['x2'])

            self.prompts.append(prompt)
            self.targets.append(target)

    def _build_prompt(self, prices, ambient_temperatures, peak_power_limit, T_room_min, T_room_max):
        s = "[Prices]\n" + "\n".join(f"Slot {i}: {p:.3f} $/kWh" for i, p in enumerate(prices))
        s += "\n\n[Ambient Temperatures]\n" + "\n".join(f"Slot {i}: {t:.1f}°C" for i, t in enumerate(ambient_temperatures))
        s += "\n\n[AC Units]\n"
        for i, (power, t_init, t_min, t_max) in enumerate(zip(
            self.P_AC, self.T_room_init, T_room_min, T_room_max), start=1):
            s += (f"AC Unit {i}:\n"
                  f"  Power: {power:.1f} kW\n"
                  f"  Initial Temp: {t_init:.1f}°C\n"
                  f"  Min Temp: {t_min:.1f}°C\n"
                  f"  Max Temp: {t_max:.1f}°C\n")
        s += f"\n[Peak Limit]\n{peak_power_limit:.1f} kW\n"
        s += "\n[Thermal Parameters]\n"
        s += f"Alpha: {self.alpha}\nG: {self.G}\n"
        s += "\n[Objective]\nMinimize total electricity cost while maintaining comfortable temperatures.\n"
        s += f"\n[Output]\nProvide two {self.N+1}-length schedules for each AC unit (comma-separated values for each, separated by semicolon)."
        return s

    def _schedules_to_text(self, x1, x2):
        """Convert schedules to text format (comma-separated values)"""
        x1_str = ",".join([f"{x:.2f}" for x in x1])
        x2_str = ",".join([f"{x:.2f}" for x in x2])
        return f"{x1_str};{x2_str}"

    def __len__(self):
        return len(self.prompts)

    def __getitem__(self, idx):
        return self.prompts[idx], self.targets[idx]

# 3) Model: T5 for text-to-text generation
class T5ScheduleGenerator(nn.Module):
    def __init__(self):
        super().__init__()
        self.tokenizer = T5Tokenizer.from_pretrained("t5-small")
        self.model = T5ForConditionalGeneration.from_pretrained("t5-small")

    def forward(self, input_ids, attention_mask, labels=None):
        return self.model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )

# 4) Training with T5 (now with loss tracking)
def train_model(model, dataset, epochs=20, batch_size=8, lr=3e-5, save_dir="t5_tcl_model"):
    device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
    model.to(device)

    def collate_fn(batch):
        prompts, targets = zip(*batch)
        tokenizer = model.tokenizer
        inputs = tokenizer(
            list(prompts),
            padding=True,
            truncation=True,
            max_length=512,
            return_tensors="pt"
        )
        labels = tokenizer(
            list(targets),
            padding=True,
            truncation=True,
            max_length=512,
            return_tensors="pt"
        ).input_ids

        labels[labels == tokenizer.pad_token_id] = -100

        return {
            "input_ids": inputs.input_ids,
            "attention_mask": inputs.attention_mask,
            "labels": labels
        }

    loader = DataLoader(dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)

    optim = AdamW(model.parameters(), lr=lr)
    total_steps = len(loader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=0,
        num_training_steps=total_steps
    )

    # Store losses for plotting
    epoch_losses = []
    batch_losses = []

    for ep in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        for batch in loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            loss = outputs.loss

            optim.zero_grad()
            loss.backward()
            optim.step()
            scheduler.step()

            running_loss += loss.item() * batch["input_ids"].size(0)
            batch_losses.append(loss.item())  # Store batch loss

        avg_loss = running_loss / len(dataset)
        epoch_losses.append(avg_loss)  # Store epoch loss
        print(f"Epoch {ep}/{epochs} — Avg Loss: {avg_loss:.6f}")

    # Plot the training loss
    plt.figure(figsize=(12, 5))

    # Plot epoch losses
    plt.subplot(1, 2, 1)
    plt.plot(range(1, epochs+1), epoch_losses, 'b-')
    plt.title('Epoch Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)

    # Plot batch losses (smoothed)
    plt.subplot(1, 2, 2)
    window_size = max(len(batch_losses) // 100, 1)  # Adjust window size for smoothing
    smoothed_losses = np.convolve(batch_losses, np.ones(window_size)/window_size, mode='valid')
    plt.plot(smoothed_losses, 'r-')
    plt.title('Smoothed Batch Loss')
    plt.xlabel('Batch')
    plt.ylabel('Loss')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Save model and tokenizer
    os.makedirs(save_dir, exist_ok=True)
    model.model.save_pretrained(save_dir)
    model.tokenizer.save_pretrained(save_dir)
    print(f"Model and tokenizer saved to ./{save_dir}")

    return model

# 5) Test & plot with T5 generation
def test_and_plot(model, tests=3, N=25):
    # Fixed parameters
    P_AC = np.array([2.0, 2.5])  # kW
    T_room_init = np.array([25.0, 24.0])
    # Define ranges for min and max temperatures
    T_room_min_range = [(20.0, 23.0), (19.0, 22.0)]  # (AC1 range, AC2 range)
    T_room_max_range = [(26.0, 29.0), (25.0, 28.0)]  # (AC1 range, AC2 range)
    alpha = np.array([0.1, 0.1])
    G = np.array([-0.3, -0.3])
    peak_range = (3.0, 6.0)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    tokenizer = model.tokenizer
    model.eval().to(device)

    for t_i in range(1, tests + 1):
        # Generate test case
        prices = 0.05 + (0.2 - 0.05) * np.random.rand(N+1)
        ambient_temperatures = 28 + (35 - 28) * np.random.rand(N+1)
        peak_power_limit = np.random.uniform(*peak_range)

        # Randomly sample min and max temperatures for this test
        T_room_min = [
            np.random.uniform(*T_room_min_range[0]),
            np.random.uniform(*T_room_min_range[1])
        ]
        T_room_max = [
            np.random.uniform(*T_room_max_range[0]),
            np.random.uniform(*T_room_max_range[1])
        ]

        # Build prompt
        p = "[Prices]\n" + "\n".join(f"Slot {i}: {v:.3f} $/kWh" for i, v in enumerate(prices))
        p += "\n\n[Ambient Temperatures]\n" + "\n".join(f"Slot {i}: {t:.1f}°C" for i, t in enumerate(ambient_temperatures))
        p += "\n\n[AC Units]\n"
        for i, (power, t_init, t_min, t_max) in enumerate(zip(
            P_AC, T_room_init, T_room_min, T_room_max), start=1):
            p += (f"AC Unit {i}:\n"
                  f"  Power: {power:.1f} kW\n"
                  f"  Initial Temp: {t_init:.1f}°C\n"
                  f"  Min Temp: {t_min:.1f}°C\n"
                  f"  Max Temp: {t_max:.1f}°C\n")
        p += f"\n[Peak Limit]\n{peak_power_limit:.1f} kW\n"
        p += "\n[Thermal Parameters]\n"
        p += f"Alpha: {alpha}\nG: {G}\n"
        p += "\n[Objective]\nMinimize total electricity cost while maintaining comfortable temperatures.\n"
        p += f"\n[Output]\nProvide two {N+1}-length schedules for each AC unit (comma-separated values for each, separated by semicolon)."

        # Generate prediction with T5
        with torch.no_grad():
            inputs = tokenizer(
                [p],
                padding=True,
                truncation=True,
                max_length=512,
                return_tensors="pt"
            ).to(device)

            outputs = model.model.generate(
                input_ids=inputs.input_ids,
                attention_mask=inputs.attention_mask,
                max_length=512,
                num_beams=5,
                early_stopping=True
            )

            pred_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Parse the predicted text back to schedules
        try:
            schedules = pred_text.split(";")
            pred_x1 = np.array([float(x) for x in schedules[0].split(",")])
            pred_x2 = np.array([float(x) for x in schedules[1].split(",")])
        except:
            print(f"Failed to parse model output: {pred_text}")
            pred_x1 = np.zeros(N+1)
            pred_x2 = np.zeros(N+1)

        # Get optimal solution for comparison
        optimal = generate_optimal_schedule(
            prices, ambient_temperatures,
            P_AC, T_room_init,
            T_room_min, T_room_max,
            alpha, G, peak_power_limit
        )

        # Calculate predicted temperature profiles
        pred_T1 = np.zeros(N+1)
        pred_T2 = np.zeros(N+1)
        pred_T1[0] = T_room_init[0]
        pred_T2[0] = T_room_init[1]

        for n in range(N):
            pred_T1[n+1] = pred_T1[n] + alpha[0]*(ambient_temperatures[n] - pred_T1[n]) + G[0]*P_AC[0]*pred_x1[n]
            pred_T2[n+1] = pred_T2[n] + alpha[1]*(ambient_temperatures[n] - pred_T2[n]) + G[1]*P_AC[1]*pred_x2[n]

        # Plot predicted schedule
        hours = np.arange(N+1)
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

        # Power consumption
        ax1.bar(hours, pred_x1*P_AC[0], label="AC Unit 1", alpha=0.7)
        ax1.bar(hours, pred_x2*P_AC[1], bottom=pred_x1*P_AC[0], label="AC Unit 2", alpha=0.7)
        ax1.axhline(peak_power_limit, color='r', linestyle='--', label="Peak Limit")
        ax1.set_title(f"Predicted AC Schedule (Test {t_i})")
        ax1.set_ylabel("Power (kW)")
        ax1.legend()
        ax1.grid(True)

        # Temperatures
        ax2.plot(hours, ambient_temperatures, 'g-', label="Ambient Temp")
        ax2.plot(hours, pred_T1, 'b-', label="Predicted Temp AC1")
        ax2.plot(hours, pred_T2, 'r-', label="Predicted Temp AC2")
        ax2.axhline(T_room_min[0], color='b', linestyle=':', label="Min/Max Temp")
        ax2.axhline(T_room_max[0], color='b', linestyle=':')
        ax2.axhline(T_room_min[1], color='r', linestyle=':')
        ax2.axhline(T_room_max[1], color='r', linestyle=':')
        ax2.set_title("Predicted Temperature Profiles")
        ax2.set_xlabel("Hour")
        ax2.set_ylabel("Temperature (°C)")
        ax2.legend()
        ax2.grid(True)

        plt.tight_layout()
        plt.show()

        # Plot optimal schedule for comparison
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

        # Power consumption
        ax1.bar(hours, optimal['x1']*P_AC[0], label="AC Unit 1", alpha=0.7)
        ax1.bar(hours, optimal['x2']*P_AC[1], bottom=optimal['x1']*P_AC[0], label="AC Unit 2", alpha=0.7)
        ax1.axhline(peak_power_limit, color='r', linestyle='--', label="Peak Limit")
        ax1.set_title(f"Optimal AC Schedule (Test {t_i})")
        ax1.set_ylabel("Power (kW)")
        ax1.legend()
        ax1.grid(True)

        # Temperatures (optimal would need to be calculated similarly)
        opt_T1 = np.zeros(N+1)
        opt_T2 = np.zeros(N+1)
        opt_T1[0] = T_room_init[0]
        opt_T2[0] = T_room_init[1]
        for n in range(N):
            opt_T1[n+1] = opt_T1[n] + alpha[0]*(ambient_temperatures[n] - opt_T1[n]) + G[0]*P_AC[0]*optimal['x1'][n]
            opt_T2[n+1] = opt_T2[n] + alpha[1]*(ambient_temperatures[n] - opt_T2[n]) + G[1]*P_AC[1]*optimal['x2'][n]

        ax2.plot(hours, ambient_temperatures, 'g-', label="Ambient Temp")
        ax2.plot(hours, opt_T1, 'b-', label="Optimal Temp AC1")
        ax2.plot(hours, opt_T2, 'r-', label="Optimal Temp AC2")
        ax2.axhline(T_room_min[0], color='b', linestyle=':', label="Min/Max Temp")
        ax2.axhline(T_room_max[0], color='b', linestyle=':')
        ax2.axhline(T_room_min[1], color='r', linestyle=':')
        ax2.axhline(T_room_max[1], color='r', linestyle=':')
        ax2.set_title("Optimal Temperature Profiles")
        ax2.set_xlabel("Hour")
        ax2.set_ylabel("Temperature (°C)")
        ax2.legend()
        ax2.grid(True)

        plt.tight_layout()
        plt.show()

        # Print cost summary
        pred_cost = np.sum(prices * (pred_x1*P_AC[0] + pred_x2*P_AC[1]))
        opt_cost = np.sum(prices * (optimal['x1']*P_AC[0] + optimal['x2']*P_AC[1]))

        print(f"Test {t_i}: Pred Cost=${pred_cost:.2f}, Opt Cost=${opt_cost:.2f}, Gap=${pred_cost - opt_cost:.2f}\n")
        print(f"Generated output:\n{pred_text}\n")

# 6) Entry point
if __name__ == "__main__":
    torch.cuda.empty_cache()
    N = 25  # Time periods (0 to N inclusive)

    # Create dataset and model
    ds = TCLSupervisedDatasetStructured(7000, N)
    model = T5ScheduleGenerator()

    # Train model (now with loss tracking and plotting)
    model = train_model(model, ds, epochs=100, batch_size=8, lr=2e-3, save_dir="t5_tcl_model")

    # Test and plot results
    test_and_plot(model, tests=5, N=N)
```
